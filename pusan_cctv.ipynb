```python
import pandas as pd
```

```python
p_cctv = pd.read_csv('./pusan_cctv.csv', encoding='euc-kr')
```

```python
p_cctv.head()
```

```python
p_cctv.shape
```

```python
tmp = p_cctv.소재지지번주소.str.split(' ')
```

```python
tmp.str.get(1)
```

```python
p_cctv['자치구'] = tmp.str.get(1)
```

```python
tmp = p_cctv.groupby('자치구')['카메라대수'].sum()
```

```python
p_cctv = tmp.reset_index()
```

```python
p_cctv.head()
```

```python
p_pop = pd.read_csv('./pusan_ingu.csv', encoding='euc-kr')
```

```python
p_pop.shape
```

```python
# 컬럼명 변경
dic = { 
    p_pop.columns[0]:'자치구',
    p_pop.columns[2]:'세대수',
    p_pop.columns[3]:'인구수',
    p_pop.columns[4]:'남자',
    p_pop.columns[5]:'여자'
     }
p_pop.rename( columns=dic, inplace=True )
```

```python
p_pop = p_pop.iloc[:, [0,2,3,4,5]]
```

```python
p_pop.head()
```

```python
result = pd.merge( p_cctv, p_pop, on='자치구')
```

```python
result
```

```python
result.set_index('자치구', inplace=True)
```

```python
fp1 = np.polyfit( result['인구수'], result['카메라대수'], 1 )
fp1
```

```python
# 함수 생성
f1 = np.poly1d( fp1 )
x  = np.linspace( 50000, 400000, 100 )
```

```python
# 시각화
plt.scatter( result['인구수'], result['카메라대수'], s=50 )
# 인구대비 CCTV의 관계를 1차함수로 표현한 선
plt.plot( x, f1(x), ls='dashed', lw=3, color='g' )
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.show()
```

```python
result['오차'] = np.abs( result['카메라대수'] 
                         - f1(result['인구수']) )
df_sort = result.sort_values( by='오차', ascending=False )
df_sort
```

```python
plt.scatter( result['인구수'], result['카메라대수'], s=50 )
plt.plot( x, f1(x), ls='dashed', lw=3, color='g' )
for n in range(5):
    plt.text( df_sort['인구수'][n]*1.05, df_sort['카메라대수'][n]*0.95 ,
            df_sort.index[n], fontSize=14)
plt.xlabel('인구수')
plt.ylabel('CCTV')
plt.grid()
plt.colorbar()
plt.show()
```
